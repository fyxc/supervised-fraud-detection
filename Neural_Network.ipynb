```python
import pandas as pd
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as skl
from sklearn import preprocessing
%matplotlib inline
```

```python
training = pd.read_csv('training_data.csv')
testing = pd.read_csv('testing_data.csv')
oot = pd.read_csv('oot_data.csv')
```

```python
training.describe().transpose()
```

```python
X = training.drop('Fraud',axis=1)
y = training['Fraud']
X_test = testing.drop('Fraud',axis=1)
y_test = testing['Fraud']
```

```python
index = list(training.columns.values)
oot_new=oot[index]
X_oot = oot_new.drop('Fraud',axis=1)
y_oot = oot_new['Fraud']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
StandardScaler(copy=True, with_mean=True, with_std=True)
```

```python
X = scaler.transform(X)
X_test = scaler.transform(X_test)
```

```python
X_oot = scaler.transform(X_oot)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp = MLPClassifier(hidden_layer_sizes=[5,5])
```

```python
mlp.fit(X,y)
```

```python
predictions = mlp.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
```
